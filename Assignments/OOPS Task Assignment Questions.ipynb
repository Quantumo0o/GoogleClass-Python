Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.

Multiple inheritance is a feature in which a class can inherit attributes and methods from more than one parent class. This allows the child class to combine functionalities from multiple sources.

```python
class Engine:
    def engine_info(self):
        return "This is a V8 Engine."

class Body:
    def body_info(self):
        return "This is a Sedan body type."

# Child class inheriting from both Engine and Body
class SportCar(Engine, Body):
    def car_type(self):
        return "This is a high-performance sports car."

# Demonstrate multiple inheritance
my_sport_car = SportCar()
print(my_sport_car.engine_info())
print(my_sport_car.body_info())
print(my_sport_car.car_type())
```

Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this
class.

Getters and Setters are methods used to access and modify the private data members of a class, ensuring data encapsulation and validation. In Python, the most common way to implement these is using the @property decorator.

```python
class Student:
    def __init__(self, name, age):
        self._name = name  # Private attribute (conventionally)
        self._age = age

    # Getter method
    @property
    def age(self):
        return self._age

    # Setter method
    @age.setter
    def age(self, value):
        if value > 0:
            self._age = value
        else:
            print("Age cannot be negative.")

# Example usage:
std = Student("Alice", 20)
print(f"Current age: {std.age}") # Calls the getter

std.age = 25 # Calls the setter
print(f"Updated age: {std.age}")
```

Q5.What is method overriding in python? Write a python code to demonstrate method overriding.

Method Overriding occurs when a child class provides a specific implementation for a method that is already defined in its parent class. The version in the child class "overrides" the one in the parent.

```python
class Animal:
    def make_sound(self):
        return "Generic animal sound"

class Dog(Animal):
    # Overriding the parent method
    def make_sound(self):
        return "Bark"

class Cat(Animal):
    # Overriding the parent method
    def make_sound(self):
        return "Meow"

# Demonstrate method overriding
dog = Dog()
cat = Cat()

print(f"Dog says: {dog.make_sound()}")
print(f"Cat says: {cat.make_sound()}")
```
